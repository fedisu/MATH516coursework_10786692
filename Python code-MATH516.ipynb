```python
'''
     Import required libraries
'''
import pandas as pd  # for data manipulation and analysis
```

```python
'''
     load 'CW.csv' dataset
'''
df=pd.read_csv('CW.csv')
```

```python
'''
     print the number of samples and number of features in the dataset   
'''
numSamples=df.shape[0]
numFeatures=df.shape[1]
print(numSamples, numFeatures)
df.shape
```

```python
'''
     how many samples are Negative and Positive
'''
df['DXCURREN'].value_counts()
```

```python
'''
    Check for missing values
'''
df.isnull().sum()
```

```python
'''
    mean for each feature for Negative samples ('0')
'''
negative_stats=df[df['DXCURREN']==0].describe().T # stats
negative_stats=negative_stats[:-1] # drop last row -- which is the class label stats
negative_means=negative_stats['mean']
negative_means
```

```python
'''
    calculate the mean for each feature for Positive samples ('1')
'''
positive_stats=df[df['DXCURREN']==1].describe().T # stats
positive_stats=positive_stats[:-1] # drop last row -- which is the class label stats
positive_means=positive_stats['mean']
positive_means
```

```python
'''
    compare means for Positive and Negative samples
'''
# create empty dataframe
feat_means=pd.DataFrame()

# add a column named 'features' 
feat_means['features']=negative_stats.index

# add a column for 'neg_means'
feat_means['neg_means']=negative_means.values

# add a column for 'neg_means'
feat_means['pos_means']=positive_means.values

# calculate % fold change between positive and negative
fc=((feat_means['pos_means'] - feat_means['neg_means'])/feat_means['neg_means'])*100

# add fold change column
feat_means['fold_change']=fc

feat_means
```

```python
'''
    Calculate pvalue
'''
from scipy import stats

# copy data
dataset=df

# seperate the two sets of data
a=dataset[dataset['DXCURREN']==0].drop(columns=['DXCURREN'])
b=dataset[dataset['DXCURREN']==1].drop(columns=['DXCURREN'])

# Calculate pvalues using t-test
pvals=(pd.DataFrame(stats.ttest_ind(a,b)).T).drop(columns=[0]).rename(columns={1: "p_value"})
pvals
```

```python
'''
    rank features according to p-value
'''
# import plotting library
from matplotlib import pyplot as plt

# create a dataframe 
ranked_pvals=pd.DataFrame()

# add features column
ranked_pvals['feature'] = negative_stats.index

# add pvalue
ranked_pvals['p_value'] = pvals['p_value']

# rank dataframe by pvalue
ranked_pvals=ranked_pvals.sort_values(by=['p_value'])

# plot ranking
ranked_pvals.plot.bar(x='feature', y='p_value', rot=0,figsize=(22,8), color='green')
plt.xticks(rotation='vertical',fontsize=11)
plt.yticks(fontsize=14)
plt.xlabel('')
plt.ylabel('p_value',fontsize=14)
plt.legend(fontsize=14)
plt.title('Features Ranked by pvalue',fontsize=16)
plt.axhline(y = 0.05, color = 'r', linestyle = '--')
plt.tight_layout()

ranked_pvals
```

```python
''' 
    identify significant features
'''
significant_features = ranked_pvals[ranked_pvals.p_value <= 0.05]
significant_features.feature
```

```python
'''
    Indentify insignificant features
'''
insig_feat = ranked_pvals[ranked_pvals.p_value > 0.05]
print("The insignificant features of the dataset are:")
insig_feat.feature # only 1 insignificant feature
```

```python
'''
    Normalsise Data 
'''
df_min_max_scaled = df.copy()
  
# apply normalization techniques
for column in df_min_max_scaled.columns:
    df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())    
  
# view normalized data
df_min_max_scaled
```

```python
'''
    Data Augmentation
'''
# load the data
dataset=pd.read_csv('CW.csv')

# Plot distribution of
dataset['DXCURREN'].value_counts().plot(kind='bar', title='Count without augmentation');
```

```python
'''
    Splitting data into training and test sets
'''
from sklearn.model_selection import train_test_split

df = df_min_max_scaled.copy()

# remove class column and insignificant features from training set
x = df.drop(columns = ['DXCURREN', 'EICV', 'RID'])
y = df['DXCURREN']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, 
                                                   random_state = 42)
```

```python
'''
    Logistic Regression Classifier 
'''
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression

lr_class = LogisticRegression(random_state = 42)

lr_class.fit(x_train, y_train)

lr_pred = lr_class.predict(x_test)

lr_cfm = confusion_matrix(y_test, lr_pred)

lr_TP=lr_cfm[1,1]
lr_TN=lr_cfm[0,0]
lr_FP=lr_cfm[1,0]
lr_FN=lr_cfm[0,1]

lr_Accuracy = (lr_TP+lr_TN)/(lr_TP+lr_TN+lr_FN+lr_FP)
lr_Sensitivity = lr_TP / (lr_TP + lr_FN)
lr_Specificity = lr_TN / (lr_TN + lr_FP)
lr_Precision = lr_TP / (lr_TP + lr_FP)
lr_Recall = lr_TP / (lr_TP + lr_FN)

print("Accuracy =",lr_Accuracy,", Sensitivity =",lr_Sensitivity,
      ", Specificity =",lr_Specificity, ", Precision =",lr_Precision, 
      ", Recall =",lr_Recall)

ConfusionMatrixDisplay(lr_cfm).plot() # plot confusion matrix
```

```python
'''
    Random Forest Classifier 
'''
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(n_estimators = 100, random_state = 42)

rf_classifier.fit(x_train, y_train)

rf_pred = rf_classifier.predict(x_test)

rf_cfm = confusion_matrix(y_test, rf_pred)

rf_TP=rf_cfm[1,1]
rf_TN=rf_cfm[0,0]
rf_FP=rf_cfm[1,0]
rf_FN=rf_cfm[0,1]

rf_Accuracy = (rf_TP+rf_TN)/(rf_TP+rf_TN+rf_FN+rf_FP)
rf_Sensitivity = rf_TP / (rf_TP + rf_FN)
rf_Specificity = rf_TN / (rf_TN + rf_FP)
rf_Precision = rf_TP / (rf_TP + rf_FP)
rf_Recall = rf_TP / (rf_TP + rf_FN)

print("Accuracy =",rf_Accuracy,", Sensitivity =",rf_Sensitivity,
      ", Specificity =",rf_Specificity, ", Precision =",rf_Precision, 
      ", Recall =",rf_Recall)

ConfusionMatrixDisplay(rf_cfm).plot() # plot confusion matrix
```

```python
'''
    Compare the model performances
'''
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

# Classifier
classif = ['LogisticRegression', 'RandomForestClassifier']

# Accuracy
lr_acc = accuracy_score(y_test, lr_pred)*100
rf_acc = accuracy_score(y_test, rf_pred)*100
accuracy = [lr_acc, rf_acc]

# F1 score
lr_f1 = f1_score(y_test, lr_pred)
rf_f1 = f1_score(y_test, rf_pred)
score_f1 = [lr_f1, rf_f1]

# Precision
lr_prec = precision_score(y_test, lr_pred)
rf_prec = precision_score(y_test, rf_pred)
precision = [lr_prec, rf_prec]

# Recall
lr_rec = recall_score(y_test, lr_pred)
rf_rec = recall_score(y_test, rf_pred)
recall = [lr_rec, rf_rec]

# Sensitivity
sensitivity = [lr_Sensitivity, rf_Sensitivity]

# Specificity
specificity = [lr_Specificity, rf_Specificity]

class_performance = pd.DataFrame()
class_performance['classifiers'] = classif
class_performance['accuracy'] = accuracy
class_performance['f1'] = score_f1
class_performance['precision'] = precision
class_performance['recall'] = recall
class_performance['sensitivity'] = sensitivity
class_performance['specificity'] = specificity

class_performance
```

```python
'''
    ROC curve Logistic Regression
'''
import matplotlib.pyplot as plt
from sklearn import metrics

y_pred_prob = lr_class.predict_proba(x_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_prob)
auc_lr = metrics.roc_auc_score(y_test, y_pred_prob)

plt.plot(fpr,tpr, label = "AUC = " + str(auc_lr))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.title('ROC Curve for Logistic Regression Model', fontsize = 12)
plt.show()
```

```python
'''
    ROC curve Random Forest Classifier
'''
y_pred_prob = rf_classifier.predict_proba(x_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_prob)
auc_rf = metrics.roc_auc_score(y_test, y_pred_prob)

plt.plot(fpr,tpr, label = "AUC = " + str(auc_rf))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.title('ROC Curve for Random Forest Classifier', fontsize = 12)
plt.show()
```

```python
'''
    Comparing the AUC of each model
'''
lr_auc = auc_lr*100
rf_auc = auc_rf*100
print(lr_auc, rf_auc)
d = lr_auc - rf_auc
diff = d.round(2)
print("Logisitic Regression model is greater by",diff,"%")
```
